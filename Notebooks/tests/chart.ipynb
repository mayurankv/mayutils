```python
from mayutils.visualisation.graphs import (
    MainAxisConfig,
    MainAxisConfigs,
)

x = np.linspace(-5, 80, 10)
y = np.linspace(-5, 60, 10)
xGrid, yGrid = np.meshgrid(y, x)
z = xGrid**3 + yGrid**3

SubPlot(
    "",
    subplot_config=SubPlotConfig(
        plots=(
            (
                PlotConfig(
                    yaxes_configs=(
                        TracesConfig.from_trace(go.Scatter(x=[1, 2], y=[1, 2])),
                        TracesConfig.from_trace(go.Scatter(x=[1, 2], y=[1, 3])),
                        TracesConfig.from_trace(go.Scatter(x=[1, 3], y=[1, 3])),
                        TracesConfig.from_trace(go.Scatter(x=[1, 3], y=[1, 3])),
                        TracesConfig.from_trace(go.Scatter(x=[1, 3], y=[1, 3])),
                        TracesConfig.from_trace(go.Scatter(x=[1, 3], y=[1, 3])),
                    ),
                    xaxis_config=dict(),
                ),
                PlotConfig.from_trace(go.Scatter(x=[1, 2], y=[1, 2])),
            ),
            (
                PlotConfig.from_trace(go.Surface(x=x, y=y, z=z, showscale=True)),
                PlotConfig.from_trace(go.Scatter(x=[1, 2], y=[1, 2])),
            ),
        ),
        main_axis_configs=MainAxisConfigs(
            xaxis=MainAxisConfig(
                config=dict(),
            ),
            yaxes=(
                MainAxisConfig(),
                MainAxisConfig(),
                MainAxisConfig(),
                MainAxisConfig(),
                MainAxisConfig(),
            ),
        ),
    ),
    layout=dict(
        xaxis=dict(
            rangeslider=dict(
                visible=True,
                yaxis=dict(
                    range=[0, 0.2],
                ),
            )
        )
    ),
)
```

```python
import numpy as np

from mayutils.visualisation.graphs import (
    Bar3d,
    PlotConfig,
    TracesConfig,
)

x = np.array(["C", "A", "z", "x"] + list(np.arange(6)))
y = np.array(["F", "1", "2"])

np.random.seed(29)

df = DataFrame(
    columns=y,
    index=x,
    data=1 + np.arange(len(y) * len(x)).reshape(len(x), len(y)) / 10,
)
df.at["A", "1"] = np.nan

Plot(
    "",
    PlotConfig(
        yaxes_configs=(
            TracesConfig(
                (
                    Bar3d.from_dataframe(
                        df=df,
                        value_weights=True,
                        name="foo",
                        showlegend=True,
                        x_mapping=sorted(df.index),
                        y_mapping=sorted(df.columns),
                    ),
                ),
            ),
            TracesConfig(
                (
                    Bar3d.from_dataframe(
                        df=df.T,
                        value_weights=True,
                        visible="legendonly",
                        name="bar",
                        showlegend=True,
                        y_mapping=sorted(df.index),
                        x_mapping=sorted(df.columns),
                    ),
                ),
            ),
        )
    ),
    layout=dict(),
)

# TODO: Categorical Axis
```
