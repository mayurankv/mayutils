```python
from mayutils.visualisation.graphs.plotly.charts import MainAxisConfig, MainAxisConfigs

x = np.linspace(-5, 80, 10)
y = np.linspace(-5, 60, 10)
xGrid, yGrid = np.meshgrid(y, x)
z = xGrid**3 + yGrid**3

SubPlot(
    "",
    subplot_config=SubPlotConfig(
        plots=(
            (
                PlotConfig(
                    yaxes_configs=(
                        TracesConfig.from_trace(go.Scatter(x=[1, 2], y=[1, 2])),
                        TracesConfig.from_trace(go.Scatter(x=[1, 2], y=[1, 3])),
                        TracesConfig.from_trace(go.Scatter(x=[1, 3], y=[1, 3])),
                        TracesConfig.from_trace(go.Scatter(x=[1, 3], y=[1, 3])),
                        TracesConfig.from_trace(go.Scatter(x=[1, 3], y=[1, 3])),
                        TracesConfig.from_trace(go.Scatter(x=[1, 3], y=[1, 3])),
                    ),
                    xaxis_config=dict(),
                ),
                PlotConfig.from_trace(go.Scatter(x=[1, 2], y=[1, 2])),
            ),
            (
                PlotConfig.from_trace(go.Surface(x=x, y=y, z=z, showscale=True)),
                PlotConfig.from_trace(go.Scatter(x=[1, 2], y=[1, 2])),
            ),
        ),
        main_axis_configs=MainAxisConfigs(
            xaxis=MainAxisConfig(
                config=dict(),
            ),
            yaxes=(
                MainAxisConfig(),
                MainAxisConfig(),
                MainAxisConfig(),
                MainAxisConfig(),
                MainAxisConfig(),
            ),
        ),
    ),
    layout=dict(
        xaxis=dict(
            rangeslider=dict(
                visible=True,
                yaxis=dict(
                    range=[0, 0.2],
                ),
            )
        )
    ),
)
```

```python
import numpy as np
from pandas import DataFrame

from mayutils.visualisation.graphs.plotly.charts import PlotConfig, TracesConfig
from mayutils.visualisation.graphs.plotly.traces import Bar3d

x = np.array(["C", "A", "z", "x"] + list(np.arange(6)))
y = np.array(["F", "1", "2"])

np.random.seed(29)

df = DataFrame(
    columns=y,
    index=x,
    data=1 + np.arange(len(y) * len(x)).reshape(len(x), len(y)) / 10,
)
df.at["A", "1"] = np.nan

Plot(
    "",
    PlotConfig(
        yaxes_configs=(
            TracesConfig(
                (
                    Bar3d.from_dataframe(
                        df=df,
                        value_weights=True,
                        name="foo",
                        showlegend=True,
                        x_mapping=sorted(df.index),
                        y_mapping=sorted(df.columns),
                    ),
                ),
            ),
            TracesConfig(
                (
                    Bar3d.from_dataframe(
                        df=df.T,
                        value_weights=True,
                        visible="legendonly",
                        name="bar",
                        showlegend=True,
                        y_mapping=sorted(df.index),
                        x_mapping=sorted(df.columns),
                    ),
                ),
            ),
        )
    ),
    layout=dict(),
)

# TODO: Categorical Axis
```

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Simulate data
x = np.linspace(0, 10, 100)
y_data = [np.sin(x + i) for i in range(4)]

# Create 2x2 subplot layout
fig = make_subplots(rows=2, cols=2)

# Add traces
for i in range(4):
    fig.add_trace(
        go.Scatter(x=x, y=y_data[i], name=f"Trace {i}"), row=i // 2 + 1, col=i % 2 + 1
    )

# Predefine axis IDs
axes = ["", "2", "3", "4"]
xaxis_ids = [f"xaxis{a}" for a in axes]
yaxis_ids = [f"yaxis{a}" for a in axes]

# Button to zoom into each subplot
buttons = []

for i in range(4):
    vis = [False] * 4
    vis[i] = True

    # layout_update = {ax: dict(domain=[0, 0.0001]) for ax in xaxis_ids + yaxis_ids}
    # layout_update[xaxis_ids[i]] = dict(domain=[0, 1])
    # layout_update[yaxis_ids[i]] = dict(domain=[0, 1])

    buttons.append(
        dict(
            label=f"Zoom Trace {i}",
            method="update",
            args=[
                {"visible": vis},
                # layout_update,
            ],
        )
    )

# "Show All" button
buttons.insert(
    0,
    dict(
        label="Show All",
        method="update",
        args=[
            {"visible": [True] * 4},
            # {  # Reset to 2x2 layout
            #     "xaxis": dict(domain=[0, 0.45]),
            #     "yaxis": dict(domain=[0.55, 1]),
            #     "xaxis2": dict(domain=[0.55, 1]),
            #     "yaxis2": dict(domain=[0.55, 1]),
            #     "xaxis3": dict(domain=[0, 0.45]),
            #     "yaxis3": dict(domain=[0, 0.45]),
            #     "xaxis4": dict(domain=[0.55, 1]),
            #     "yaxis4": dict(domain=[0, 0.45]),
            # },
        ],
    ),
)

# Add buttons to layout
fig.update_layout(
    updatemenus=[dict(buttons=buttons, direction="down", showactive=False, x=1.1, y=1)],
    height=600,
    width=800,
    title="Click a button to zoom into a subplot",
    showlegend=True,
)

fig.show()
```
